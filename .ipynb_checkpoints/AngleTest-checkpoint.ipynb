```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def calculate_coordinates(dist, angle):
    x = 0.0
    z = 0.0
    
    if angle >= 360:
        angle = angle - 360
    
    if angle <= 90:
        angle = 90 - angle
        x = np.cos((np.deg2rad(angle)))
        z = np.sin((np.deg2rad(angle)))
    elif angle > 90 and angle <= 180:
        angle = angle - 90
        x = np.cos((np.deg2rad(angle)))
        z = -(np.sin((np.deg2rad(angle))))
    elif angle > 180 and angle <= 270:
        angle = 270 - angle
        x = -(np.cos((np.deg2rad(angle))))
        z = -(np.sin((np.deg2rad(angle))))
    elif angle > 270 and angle < 360:
        angle = angle - 270
        x = -(np.cos((np.deg2rad(angle))))
        z = np.sin((np.deg2rad(angle)))
        
    return x * dist, z * dist
    
```

```python
from pylab import *

for a in range(0, 361, 15):
    angle = float(a)

    x = []
    z = []
    for i in range(0, 50):
        x_, z_ = calculate_coordinates(i, angle)
        x.append(x_)
        z.append(z_)

#     plt.plot(x, z)
#     plt.xlabel('x')
#     plt.ylabel('z')
#     plt.show()
    print(angle)
    scatter(x,z)
    show()
```
