```python
def calculate_coordinates(dist, angle):
    x = 0.0
    z = 0.0
    
    if angle >= 360:
        angle = angle - 360
    
    x = np.sin((np.deg2rad(angle)))
    z = np.cos((np.deg2rad(angle)))
        
    return x * dist, z * dist
    
```

```python
from pylab import *

for a in range(0, 361, 15):
    angle = float(a)

    x = []
    z = []
    for i in range(0, 25):
        x_, z_ = calculate_coordinates(i, angle)
        x.append(x_)
        z.append(z_)

    print(angle)
    scatter(x,z)
    ylim((-50, 50))
    xlim((-50, 50))
    show()
```
